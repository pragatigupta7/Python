# date time conversion - string to datetime


    format | meaning
    %d     | day of the month 01-31
    %m     | month of the year 01-12
    %y     | year
    %H     | hour 00-23
    %M     | minute 00-59
    %S     | second 00-59

```python
from datetime import datetime
datestr = '20141013T000000'
date = datetime.strptime(datestr,'%Y%m%dT%H%M%S')
print(date,type(date))
```

# immigration data analysis

```python
df = pd.read_excel('Canada.xlsx',sheet_name = 1,skiprows = 20,skipfooter=2)
df.head(2)
```

- rename columns
- drop columns
- add column

```python
cols_to_rename = {
    'OdName':'Country',
    'AreaName': 'Continent',
    'RegName':'Region',
    'DevName': 'Status',
}
df = df.rename(columns=cols_to_rename)
df.head(2)
```

### drop columns

```python
# check no of nunique values in data
print("Type Unique values:",df.Type.nunique())
print("coverage unique values:",df.Coverage.nunique())

# counr unique values in a column
print("Coverage unique values count\n",df.Coverage.value_counts())
```

```python
cols_to_drop = ['AREA','REG','DEV','type','Coverage']
df = df.drop(columns=cols_to_drop)
df.head(2)
```

set index - not for every dataframe you use

```python
df = df.set_index('Country')
df.head()
```

make all columns to lower case-optional step

```python
df.columns = [str(name).lower() for name in df.columns.tolist()]
df.head()
```

```python
years = list(map(str,range(1980,2014)))
print(years)
```

- adding a column (total)

```python
df[years].sum(axis=1) #Axis 1 for row wise sum,default is o (column wise sum)
```

```python
df['total'] = df[years].sum(axis=1)
df.head()
```

visualization

```python
df.isnull().sum().plot(figsize=(10,1),kind='bar')   #no null value dictacted
```

total immigration visualization

```python
df.total.plot(rot=90,
              figsize=(25,5),
              xticks=range(0,df.shape[0],2),
              title='Total Immigration')
```

```python
#top 25 countries with higest immigration
df.sort_values('total',ascending=False).head(25).plot(kind='bar',y='total',figsize=(15,5),title='top 25 countries with highest immigration',color='red',alpha=.5,xlabel='Country',ylabel='Immigration')
```
